```python
def heatmap_over_image(image, data, ap_name):
    fig, ax = plt.subplots()
    ax.imshow(image)
    
    plt.scatter(x=data['x'].values, y=data['y'].values, c=data[ap_name].values, cmap='gray')
    plt.title('House Position WiFi RSSI Levels')
    plt.xlabel("x position")
    plt.ylabel("y position")
    plt.savefig("house_wifi_"+ap_name+".png")
```

```python
image = plt.imread("wifi_test_grey.PNG")
ap_name="Telekom-iU1HVB"
heatmap_over_image(image, data, ap_name)
```

```python
def save_all_images(image, data):
    for ap_name in data.columns[2:-2]:
        heatmap_over_image(image, data, ap_name)
```

```python
save_all_images(image,data)
```
